```python
import pandas as pd
import numpy as np
from sklearn import decomposition
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("career_compute_dataset.csv")
```

```python
data = dataset.iloc[:49,:-1].values
label = dataset.iloc[:49,-1].values
data
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
labelencoder = LabelEncoder()
```

```python
label = labelencoder.fit_transform(label)
label
```

```python
y=pd.DataFrame(label,columns=["role"])
X1 = pd.DataFrame(data,columns=['sslc','hsc','cgpa','school_type','no_of_miniprojects','no_of_projects','coresub_skill','aptitude_skill','problemsolving_skill','programming_skill','abstractthink_skill','design_skill','first_computer','first_program','lab_programs','ds_coding','technology_used','sympos_attend','sympos_won','extracurricular','learning_style','college_bench','clg_teachers_know','college_performence','college_skills'])
```

```python
from sklearn.preprocessing import Normalizer
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn import preprocessing 
from sklearn.metrics import accuracy_score
```

```python
X_train,X_test,y_train,y_test=train_test_split(X1,y,test_size=0.2,random_state=20)
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
y_pred = clf.predict(X_test)
print(y_pred)
```

```python
cm = confusion_matrix(y_test,y_pred)
accuracy = accuracy_score(y_test,y_pred)
len(cm[1])
```

```python
print("confusion matrics=",cm)
print("  ")
print("accuracy=",accuracy*100)
```
